```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
asks=pd.read_csv('asks.csv',index_col=False)
bids=pd.read_csv('bids.csv',index_col=False)
asks=asks.iloc[:999,1:]
bids=bids.iloc[:999,1:]
asks=asks.iloc[::-1]
df=asks.append(bids)
```

```python
df.columns=['price','amount']
df['price']=round(df['price'])
df=df.groupby('price').sum()
```

```python
g=sns.heatmap(df,cmap='YlGnBu')
```
